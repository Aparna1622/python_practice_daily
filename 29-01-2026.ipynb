Comments:<br>
these are used for readibility, debugging and maintenance<br>
Two types of comments:<br>
1. Single line comments<br>
'#' is used for single line commenting<br>
2. Multi line comments<br>
it is used for documenting functions, modules and classes<br>
''' or """ are used for this purpose

```python
#single line comment
name="Aparna"
#print the name
print(name)
```

```python
""" Multiline
strings as comments
"""
print("world is so big")
```

Literals<br>
Fixed values written directly in the code<br>
Do not change by themselves<br>
Cannot store data, they are the data<br>
dont have name<br>
used to assign values to variables

```python
#Literal
x=10 #10 is a literal
x
```

Variables<br>
named memory locations<br>
used to store data<br>
values can change during execution<br>
have a name(identifier)

```python
#variable
y="age" #y is variable and "age" is literal
y
```

Rules for python variables<br>
must start with a letter(a-z or A-Z) or underscore(_)<br>
cannot start with a number<br>
can contain letters, numbers and underscore<br>
no spaces allowed<br>
case sensitive(age,Age and AGe are different)<br>
cannot be a python keyword

```python
#defining separator function
def sep():
    print("------"*6)
```

```python
#Valid variable name
name="Aparna"
print(name)
Name="Aparna"
print(Name)
_name1="Aparna"
print(_name1)
```

```python
#Invalid variable names which give error
1name="Aparna"
print(1name)
my name="Naram"
print(my name)
def="Aparna"
print(def)
```

```python
#assigning multiple values to multiple variables
a,b,c=1,2,3
print(a)
print(b)
print(c)
```

```python
#same value to multiple variables
a=b=c=2.5
print(a)
print(b)
print(c)
```

```python
#unpack a collection
fruits=['apple','banana','cherry']
x,y,z=fruits
print(x)
print(y)
print(z)
```

Data types<br>
what type of data a variable holds<br>
what operations can be performed on it
python automatically assigns a data type when a variable is created. No need to specify the data type during variable creation

Different data types<br>
text type: str<br>
numeric type: int, float, complex<br>
sequence: list, tuple, range<br>
mapping: dict<br>
set: set, frozen set<br>
boolean: bool<br>
binary types: bytes, bytearray, memory view<br>
None: None

```python
#numeric type
z=1
print(type(z))
y=2.5
print(type(y))
x=1+2j
print(type(x))
```
