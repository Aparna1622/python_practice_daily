Operators in python

#Arithmetic operators<br>
Addition(+)<br>
Subtraction(-)<br>
Multiplication(*)<br>
Division(/)<br>
Modulus(%)<br>
Floor division(//)<br>
Exponential(**)

```python
#Example
a=12
b=3
print("addition:",a+b)
print("subtraction:",a-b)
print("multiplication:",a*b)
print("division:",a/b)
print("modulus:",a%b)
print("floor division",a//b)
print("exponential",a**b)
```

#Assignment operators<br>
Basic assignment(=)<br>
Addition assignment(+=)<br>
Subtraction assignment(-=)<br>
Multiplication assignment(*=)<br>
Floor division assignment(//=)<br>
Modulus assignment(%=)<br>
Exponential assignment(**=)

```python
x=6
print("basic assignment:",x)
x+=4
print("additional assignment:",x)
x-=2
print("subtraction assignment:",x)
x*=2
print("multiplication assignment:",x)
x/=2
print("division assignment:",x)
x%=2
print("modulus assignment:",x)
x//=2
print("floor division assignment:",x)
x**=2
print("exponential assignment:",x)
```

```python
#Comparison operators
a=10
b=20
print("Equal to:",a==b)
print("Not equal to:",a!=b)
print("Less than:",a<b)
print("Greater than:",a>b)
print("Less than or equal to:",a<=b)
print("Greater than or equal to:",a>=b)
```

```python
#Logical operators
x=True
y=False
print("Logical AND (x and y):",x and y)
print("Logical OR (x or y):",x or y)
print("Logical NOT (not x):",not x)
```

```python
#Identity operators
x=[1,2,3]
y=[1,2,3]
z=x
print("Identity is (x is z):",x is z)
print("Identity is not (x is not y):",x is not y)
```

```python
#Membership operators
x=[1,2,3]
print("Membership in list:",3 in x)
```

```python
# Bitwise operator
# Define variables
a = 60  # 0011 1100 in binary
b = 13  # 0000 1101 in binary

# Bitwise AND
print("Bitwise AND (a & b):", a & b)  # Output: 12 (0000 1100 in binary)

# Bitwise OR
print("Bitwise OR (a | b):", a | b)  # Output: 61 (0011 1101 in binary)

# Bitwise XOR
print("Bitwise XOR (a ^ b):", a ^ b)  # Output: 49 (0011 0001 in binary)

# Bitwise NOT
print("Bitwise NOT (~a):", ~a)  # Output: -61 (inverted bits of a)

# Bitwise Left Shift
print("Bitwise Left Shift (a << 2):", a << 2)  # Output: 240 (1111 0000 in binary)

# Bitwise Right Shift
print("Bitwise Right Shift (a >> 2):", a >> 2)  # Output: 15 (0000 1111 in binary)
```

Data type:<br>
describes what kind of value a variable is storing

```python
#Numeric data type
#integer
z=10
print(type(z))
#float
z=10.2
print(type(z))
#complex
z=2+4j
print(type(z))
```

```python
#string data type
name="Aparna"
print(type(name))
```

```python
#boolean data type
answer=True
print(type(answer))
```

```python
#Sequence data type- it stores multiple values
#list
#ordered and mutable
l=[1,2,3]
print(type(l))
#tuple
#ordered and immutable
a=(12,13,14)
print(type(a))
#range
a=range(4)
print(type(a))
```

```python
#set data type
#unordered and no duplicates
a={1,2,3}
print(type(a))
```

```python
#dictionary data type
#key value pair
data={
    "name":"aparna",
    "age":28
}
print(type(data))
```
